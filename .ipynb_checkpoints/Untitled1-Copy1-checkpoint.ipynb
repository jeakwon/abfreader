```python
import numpy as np
import pandas as pd
from scipy.signal import decimate
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import pyabf
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
plt.style.use('seaborn')
np.set_printoptions(precision=3)
```

```python
FILENAME = '17d22009'+'.abf'

# abf file 열기
abf=pyabf.ABF(FILENAME)

# abf file data 확인
abfHeader=pyabf.header.ABFheader(FILENAME)
abfHeader.show()
# plt.plot(abf.dataX,abf.dataY)
```

```python
def movingaverage (values, window):
    weights = np.repeat(1.0, window)/window
    sma = np.convolve(values, weights, 'valid')
    return sma
```

```python
N=100
x=abf.dataX[:-N+1]
y=movingaverage(abf.dataY,N)

for i in range(2):
    x=decimate(x,q=10,zero_phase=True)
    y=decimate(y,q=10,zero_phase=True)

print(len(x))
print(len(y))
trace_color = 'k'
base_color = 'b'
gaba_color = 'r'

trace_start = 70000
trace_end = 110000

base_start = 88000
base_end = 90000

gaba_start = 93000
gaba_end = 95000
```

```python
s=1173450*6
e=11734500
plt.plot(abf.dataX[s:e],abf.dataY[s:e])
```

```python
#시작부분 코드
import time
start_time = time.time() 
#------------------------

plt.scatter(x[trace_start:base_start],y[trace_start:base_start],alpha=0.3, s=5,c=trace_color)
plt.scatter(x[base_start:base_end],y[base_start:base_end],alpha=0.3, s=5,c=base_color)
plt.scatter(x[base_end:gaba_start],y[base_end:gaba_start],alpha=0.3, s=5,c=trace_color)
plt.scatter(x[gaba_start:gaba_end],y[gaba_start:gaba_end],alpha=0.3, s=5,c=gaba_color)
plt.scatter(x[gaba_end:trace_end],y[gaba_end:trace_end],alpha=0.3, s=5,c=trace_color)
plt.rcParams["figure.figsize"] = (10,4)
plt.show()

#종료부분 코드
print("start_time", start_time) #출력해보면, 시간형식이 사람이 읽기 힘든 일련번호형식입니다.
print("--- %s seconds ---" %(time.time() - start_time))
```
